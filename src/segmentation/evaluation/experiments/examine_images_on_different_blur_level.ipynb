```python
# load the json file of the blur scores
import json
import numpy as np

with open('/home/etaylor/code_projects/thesis/data/metadata/sharpness_per_patch_scores.json') as f:
    patches_sharpness_dict = json.load(f)
    
```

```python
from pprint import pprint

pprint(patches_sharpness_dict)
```

```python
import matplotlib.pyplot as plt

# histogram of the blur scores
sharpness_values = list(patches_sharpness_dict.values())

# Plot the histogram
plt.hist(sharpness_values, bins=20, edgecolor='black')
plt.xlabel('Sharpness Value')
plt.ylabel('Frequency')
plt.title('Histogram of Sharpness Values of all Patches')
plt.show()
```

```python
import os

# Path to the directory containing the images
image_directory = "/home/etaylor/code_projects/thesis/segments/etaylor_cannabis_patches_test_26-04-2024_15-44-44/v0.1"

# Get the list of image files in the specified directory
image_files_in_directory = os.listdir(image_directory)

# Create a new dictionary with the values of the images found in the specified directory
test_patches_sharpness_dict = {image: patches_sharpness_dict[image] for image in image_files_in_directory if image in patches_sharpness_dict}

# Output the new dictionary
print(test_patches_sharpness_dict)
```

```python
sharpness_values = list(test_patches_sharpness_dict.values())

# Plot the histogram
plt.hist(sharpness_values, bins=20, edgecolor='black')
plt.xlabel('Sharpness Value')
plt.ylabel('Frequency')
plt.title('Histogram of Sharpness Values of all Patches')
plt.show()
```

```python
# get the 10 lowerst and highest sharpness images in the dict
sorted_sharpness_dict = sorted(test_patches_sharpness_dict.items(), key=lambda x: x[1])

print("10 lowest sharpness images")
pprint(sorted_sharpness_dict[:10])

print("10 highest sharpness images")
pprint(sorted_sharpness_dict[-10:])
```

```python
# plot the images togerther in the same plot
import cv2

# Get the 10 lowest sharpness images
lowest_sharpness_images = sorted_sharpness_dict[:10]

# Get the 10 highest sharpness images
highest_sharpness_images = sorted_sharpness_dict[-10:]

# plot all lowest sharpness images together in the same plot
fig, axs = plt.subplots(2, 5, figsize=(20, 10))
fig.suptitle('10 Lowest Sharpness Images')
for i, image in enumerate(lowest_sharpness_images):
    img = cv2.imread(os.path.join(image_directory, image[0]))
    axs[i//5, i%5].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axs[i//5, i%5].set_title(f"Sharpness: {image[1]}")
    axs[i//5, i%5].axis('off')
    
    
fig, axs = plt.subplots(2, 5, figsize=(20, 10))
fig.suptitle('10 Highest Sharpness Images')
for i, image in enumerate(highest_sharpness_images):
    img = cv2.imread(os.path.join(image_directory, image[0]))
    axs[i//5, i%5].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axs[i//5, i%5].set_title(f"Sharpness: {image[1]}")
    axs[i//5, i%5].axis('off')
    
plt.show()




```
